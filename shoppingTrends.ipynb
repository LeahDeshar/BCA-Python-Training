```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
data["Gender"].value_counts()
```

```python
data["Gender"].value_counts().plot(kind = "bar")
plt.title("Plot of Gender distribution")
plt.show()
```

```python
# group the data by the "Category" column and then calculates the mean of the "Purchase Amount (USD)" column for each group.
data.groupby("Category")['Purchase Amount (USD)'].mean()
```

```python
data.groupby("Category")['Purchase Amount (USD)'].mean().sort_values(ascending = False).head(2)
```

```python
data.groupby("Category")['Purchase Amount (USD)'].mean().sort_values(ascending = False).tail(2)
```

```python
data.sort_values("Review Rating",ascending = False,inplace= True)
```


*   It sorts the DataFrame in place, meaning it modifies the original DataFrame.
*  It does not return anything (None is returned), so you typically do not assign the result to a variable.

Use Scatter plot for  the correlation analysis
It reveals the positive and negative relationship between the variables
- if one variable's change impact the another it has positive relationship else it has negative relationship
-It represents the linear,non-linear ,cluster or no apparent relationship  

```python
# Use to check if there is any relationship between the two variables
data.plot(x = "Previous Purchases",y='Review Rating',kind="scatter")
plt.title("Scatter Plot")
plt.show()
```
